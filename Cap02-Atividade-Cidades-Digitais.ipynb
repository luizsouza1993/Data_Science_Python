# Group By e Reshape

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html

```python
import numpy as np
import pandas as pd
```

```python
cidadesDigitais = pd.read_csv('cidades_digitais.csv', sep = ';', encoding = 'latin-1')
```

```python
cidadesDigitais.head()
```

```python
cidadesDigitais.count()
```

```python
del cidadesDigitais['IBGE']
```

```python
# Aplicando Group By
cidadesUF_1 = cidadesDigitais.groupby('UF')
```

```python
type(cidadesUF_1)
```

```python
cidadesUF_1.groups
```

```python
# Número de grupos
len(cidadesUF_1.groups)
```

```python
ordenaUF = cidadesUF_1.size()
```

```python
# Ordena os grupos por ordem decrescente
ordenaUF.sort_values(ascending = False)
```

```python
cidadesUF_2 = cidadesDigitais.groupby(['UF','STATUS'])
```

```python
cidadesUF_2
```

```python
# Total agrupado
totalUF = cidadesUF_2.size();
totalUF = totalUF.sort_values(ascending = False);
print(totalUF)
```

```python
cidadesDigitais.head()
```

```python
cidadesUF_3 = cidadesDigitais.set_index('STATUS')
```

```python
# Aplicando Group By
cidadesUF_3 = cidadesUF_3.groupby(level = 'STATUS')
```

```python
cidadesUF_3.head()
```

```python
cidadesUF_4 = cidadesDigitais.set_index(['STATUS','UF'])
```

```python
cidadesUF_4
```

```python
cidadesUF_5 = cidadesDigitais.set_index(['UF','STATUS'])
```

```python
cidadesUF_5 = cidadesDigitais.groupby(['UF','STATUS'])
```

```python
cidadesUF_5.sum()
```

```python
cidadesUF_5.aggregate(np.max)
```

```python
cidadesUF_5.aggregate([np.sum,np.mean,np.max])
```
