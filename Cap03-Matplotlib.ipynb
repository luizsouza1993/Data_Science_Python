```python
def cria_plot():
    x = np.random.randn(5000,6)
    (figure,axes) = plt.subplots(figsize = (15,10))
    (n,bins,patches) = axes.hist(x,12,
                                density =1,
                                histtype = "bar",
                                label = ["color1","color2","color3","color4","color5","color6",])
                    
    axes.set_title("Histograma para Distribuição Normal", fontsize = 12)
    axes.set_xlabel("Dados", fontsize = 10)
    axes.set_ylabel("Frequencia",fontsize = 10)
    axes.legend()
    plt.show()
```

```python
cria_plot()
```
